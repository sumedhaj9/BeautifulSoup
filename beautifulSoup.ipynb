# BeautifulSoup for Web Scraping for IIT faculty details

```python
import bs4 as bs
import urllib.request

s=[]
sauce=urllib.request.urlopen('http://www.iitbbs.ac.in/faculty-members.php').read()
soup=bs.BeautifulSoup(sauce,'html.parser')

table=soup.find('table')   # scraping the table tags 
table_rows=table.find_all('tr') #Fetching each row inthe table
for tr in table_rows:
    td=tr.find_all('td')
    row=[i.text.strip() for i in td] #Fetching the content of each row
    print(row)
    s.append(row)
```

 Now store the above setails in an excel sheet using the below piece of code

```python
import csv
with open(r'C:\Users\Lenovo\Desktop\iit.csv', "w") as f:
    writer = csv.writer(f)
    for item_property_list in s:
        writer.writerow(item_property_list)
```

Now an excel file is created called 'iit' which contains the information contained in each row of the table scraped.
